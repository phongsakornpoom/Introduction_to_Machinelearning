```python
from sklearn.model_selection import train_test_split#divid data train is 75 and test is 25 from train
Xtrain,Xtest,Ytrain,Ytest=train_test_split(
irisdf['data'],irisdf['target'],random_state=0)
```

```python
irisdataframe=pd.DataFrame(Xtrain,columns=irisdf.feature_names)#put the data into this column name
pd.plotting.scatter_matrix(irisdataframe,c=Ytrain,figsize=(15,15),marker='o',hist_kwds={'bins':20})
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(Xtrain,Ytrain)
```

```python
import numpy as np
predict1=np.array([[3,3,3,3]])
prediction=knn.predict(predict1)
```

```python
print('target',irisdf.target,'targetname',irisdf.target_names)
#the target tell which index is the type in target name.
```

```python
irisdf['target_names'][prediction]
#we select and tell colum target names and use the prediction which is return array to tell what type it is
```

```python
ypred=knn.predict(Xtest)
print('test set prediction:',ypred)
```

```python
#first way to check accuracy
np.mean(ypred==Ytest)#accuracy 
```

```python
#second way we use knn score library
knn.score(Xtest,Ytest)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(ypred,Ytest)
```
